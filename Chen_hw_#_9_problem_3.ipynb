$L=\frac{1}{2}m\dot{x}^{2}-\frac{1}{2}k\left(x-X_{w}-l_{0}\right)^{2}$

$x-X_{w}-l_{0}=z$

$\dot{x}-\dot{X_{w}}=\dot{z}$

$L=\frac{1}{2}m\left(\dot{z}+\dot{X_{w}}\right)^{2}-\frac{1}{2}kz^{2}$

$L=\frac{1}{2}m\left(\dot{z}+\omega A\cos\omega t\right)^{2}-\frac{1}{2}kz^{2}$

$p_{z}=\frac{\partial L}{\partial\dot{z}}=m\left(\dot{z}+\omega A\cos\omega t\right)$

Therefore, the Hamiltonian becomes

$H=p_{z}\dot{z}-L$

$H=\frac{p_{z}^{2}}{2m}-p_{z}\omega A\cos\omega t+\frac{1}{2}kz^{2}$

The Hamiltonian equation is

$\frac{dz}{dt}=\frac{\partial H}{\partial p_{z}}=\frac{p_{z}}{m}-\omega A\cos\omega t$

$\frac{dp_{z}}{dt}=-\frac{\partial H}{\partial z}=-kz$

The equation of motion is

$\frac{d^{2}z}{dt^{2}}=-\frac{kz}{m}+\omega^{2}A\sin\omega t$

The solution of z can be 

$z=C\sin\left(\omega_{0}t\right)+B\cos\left(\omega_{0}t\right)-\frac{\omega^{2}mA}{\omega^{2}m-k}\sin\omega t$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(z, t,km,o,A):
    return [z[1],-km*z[0]+o*o*A*np.sin(o*t)]
```

Assumed that,initially, the object's position and velocity is 0. However, the stretch velocity $\dot{z}$ is -1 m/s when A is 1 (m) and $\omega$ is 1 (1/s).

```python
y_0 = np.array([0.0, -1.0])
t = np.linspace(0.0, 30.0, 10000)
```

```python
plt.figure(figsize=(9,7),dpi=98)
y_sol = odeint(f_int, y_0, t, args=(4,1,1))[:, 0]
plt.plot(t, y_sol,linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
y_0 = np.array([0.0, 0.0])
t = np.linspace(0.0, 30.0, 10000)
plt.figure(figsize=(9,7),dpi=98)
y_sol = odeint(f_int, y_0, t, args=(4,1,1))[:, 0]
plt.plot(t, y_sol,linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```
