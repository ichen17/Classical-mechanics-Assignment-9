$T=\frac{1}{2}mR^{2}\dot{\theta}^{2}+\frac{1}{2}mR^{2}\sin^{2}\theta\varpi^{2}$

$V=-mgR\cos\theta$

The Lagrange of the this system can be written as

$L=T-V=\frac{1}{2}mR^{2}\dot{\theta}^{2}+\frac{1}{2}mR^{2}\sin^{2}\theta\varpi^{2}+mgR\cos\theta$

The momentum is 

$\frac{\partial L}{\partial\dot{\theta}}=mR^{2}\dot{\theta}=P_{\theta}$

The Hamiltonian is

$H=P_{\theta}\dot{\theta}-L$

$H=\frac{P_{\theta}^{2}}{2mR^{2}}-\frac{1}{2}mR^{2}\sin^{2}\theta\varpi^{2}-mgR\cos\theta$

$\frac{d\theta}{dt}=\frac{\partial H}{\partial P_{\theta}}=\frac{P_{\theta}}{mR^{2}}$

$\frac{dP_{\theta}}{dt}=-\frac{\partial H}{\partial\theta}=mR^{2}\sin\theta\cos\theta\varpi^{2}-mgR\sin\theta$

$\frac{d^{2}\theta}{dt^{2}}=\sin\theta\cos\theta\varpi^{2}-\frac{g\sin\theta}{R}$

The conservation of energy is

$\frac{1}{2}mR^{2}\dot{\theta}^{2}+\frac{1}{2}mR^{2}\sin^{2}\theta\varpi^{2}-mgR\cos\theta$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def f_int(theta, t,o,R):
    g=9.8
    return [theta[1],o*o*np.sin(theta[0])*np.cos(theta[0])-g*np.sin(theta[0])/R]
y_0 = np.array([np.pi/10, 0.0])
t = np.linspace(0.0, 10.0, 10000)
```

```python
plt.figure(figsize=(9,7),dpi=98)
y_sol = odeint(f_int, y_0, t, args=(10,1))[:, 0]
plt.plot(t, y_sol,linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,7),dpi=98)
y_sol = odeint(f_int, y_0, t, args=(0.01,1))[:, 0]
plt.plot(t, y_sol,linewidth=1)
plt.grid(True)
plt.legend()
plt.show()
```
